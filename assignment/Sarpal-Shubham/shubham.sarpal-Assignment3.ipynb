```python
# Import Necessary Modules
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
# Set up the random seed
rnd=tf.set_random_seed(111)
# Display plots inline
%matplotlib inline
# Plots Figure Size
matplotlib.rcParams['figure.figsize']=(10.0,10.0)
```

```python
train = pd.read_csv("ann.csv")
print (train.head())
train_X, train_Y = np.array(train.ix[:,0:2]), np.array(train.ix[:,2])
train_X = np.asarray([[2.067788,0.258133],[0.993994,-0.609145],[-0.690315,0.749921] ,[1.023582,0.529003],[0.700747,-0.496724]]) 
train_Y = np.asarray([[1],[1],[0],[0],[1]])
print(train_X.shape, train_Y.shape) 
plt.scatter(train_X[:,0], train_X[:,1], s=50, c=train_Y, cmap=plt.cm.BuGn)
x = tf.placeholder("float",[None, 2]) 
y = tf.placeholder("float",[None, 1])
W = tf.Variable(tf.zeros([2, 1])) 
b = tf.Variable(tf.zeros([1, 1]))
activation = tf.nn.sigmoid(tf.matmul(x, W)+b) 
cost = tf.reduce_sum(tf.square(activation - y))/4 
optimizer = tf.train.GradientDescentOptimizer(.1).minimize(cost)
init = tf.initialize_all_variables()
with tf.Session() as sess: 
    sess.run(init) 
    for i in range(500): 
        error_stat= sess.run(cost, feed_dict={x: train_X, y: train_Y}) 
        print(error_stat)
        train_data = sess.run(optimizer, feed_dict={x: train_X, y: train_Y})            
        result = sess.run(activation, feed_dict={x:train_X})
        print(result)        
        #plt.scatter(train_X[:, 0], train_X[:, 1], s=50,  c=train_Y, cmap=plt.cm.BuGn)



print(train_X.shape, train_Y.shape)
plt.scatter(train_X[:,0], train_X[:,1], s=50, c=train_Y, cmap=plt.cm.BuGn)

```
